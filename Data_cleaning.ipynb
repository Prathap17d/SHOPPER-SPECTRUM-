```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset 
df=pd.read_csv('F:\Shopper Spectrum\online_retail.csv')
df.head()
```

```python
df.shape
```

```python
# Display basic information about the columns and data typesdf.info()
df.info()
```

```python
# Check for missing values in each column
df.isnull().sum()
```

```python
# change date column to right datetime format: 
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
# Drop rows where CustomerID is null
df = df.dropna(subset=['CustomerID'])

# Remove canceled orders (Invoice starts with 'C')
# We convert to string first to make sure string methods work safely
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]

# Filter out negative or zero Quantities and UnitPrices
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
```

```python
# Calculate Total Spend for each row
df['TotalSpend'] = df['Quantity'] * df['UnitPrice']
```

```python
print("True Cleaned Data Shape:", df.shape)
print("Number of Unique Customers:", df['CustomerID'].nunique())
```

```python
#find which product sells the most:
print(f"most selling prodcut: {df['Description'].value_counts().head(1)}")
print("============================")
print(f"lowest selling prodcut: {df['Description'].value_counts().tail(3)}")
```
